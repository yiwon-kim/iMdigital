# matplotlib 옵션 설정

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator, AutoMinorLocator, FuncFormatter
import seaborn as sns
import numpy as np

tips = sns.load_dataset("tips")
tips.head()
```

```python
tips.info() # 데이터 셋이 category로 되어 있는 범주형이 이상적임. object(깔끔하게 정의 되지 않은)이면 좋지 않음
```

```python
# y축 정리를 위한 사용자 지정 함수 생성 (fommatter)
def major_formatter(x, pos):
    return "%.2f$" % x # 소수점 두자리 + $ 서식
formatter = FuncFormatter(major_formatter)
```

```python
group_mean = tips.groupby(['day'])['total_bill'].agg('mean')
#group_mean
h_day = group_mean.sort_values(ascending=False).index[0]
h_mean = np.round(group_mean.sort_values(ascending=False)[0],2)
h_day, h_mean
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize=(16,5))

# 왼쪽 그래프
sns.barplot(data = tips,
            x = 'day',
            y = 'total_bill',
            errorbar=None,
            color = 'lightgray',
            alpha = 1,
            zorder = 2, # grid에 해당하는 선을 chart 뒤로 숨길 수 있는 기능, 2로 주면 뒤로 들어감
            ax = ax[0])


for p in ax[0].patches:
  print("여기 확인 필요:", p, type(p))  # 막대그래프 하나하나가 다 개별 인스턴스 각각 처리해줘야함, height : 요일별 평균
  print('높이', p.get_height(), '너비', p.get_width())
  height = np.round(p.get_height(), 2)
  fontweight = 'normal'
  color = 'k'
  if h_mean == height: # for 문에 조건문 추가 해서 조건에 맞는 애만 변경
    fontweight = "bold"
    color = 'darkred'
    p.set_facecolor(color)
    p.set_edgecolor('black')

  ax[0].text(p.get_x() + p.get_width()/2,
             height + 1,
             height, ha = 'center', size = 12,
             fontweight=fontweight, color=color) # p.get_x() x 축 위치, p.get_width()/2 폭의 절반


# 옵션
ax[0].set_title("이렇게 그리고 싶어요", size =16)
ax[0].set_ylim(-2, 30)

# spines 항목 정리
ax[0].spines['top'].set_visible(False)
ax[0].spines['left'].set_visible(False)
ax[0].spines['right'].set_visible(False)
# ax[0].spines['bottom'].set_visible(False)
ax[0].spines['left'].set_position(('outward', 20))

# y축 정리
ax[0].yaxis.set_major_locator(MultipleLocator(10)) # y축 큰 간격(major)을 10으로 세팅
ax[0].yaxis.set_major_formatter(formatter) # 20.00$
ax[0].yaxis.set_minor_locator(MultipleLocator(5)) # y축 작은 간격(minor)을 5로 세팅

# 축 라벨 지정
ax[0].set_ylabel('평균영수증($)', fontsize=14)
ax[0].set_xlabel('요일', fontsize=14)

ax[0].grid(axis= 'y', which = 'major', color = 'lightgray')
ax[0].grid(axis= 'y', which = 'minor', ls = ':')



#오른쪽 그래프
sns.barplot(data = tips,
            x = 'day',
            y = 'total_bill',
            errorbar=None,
            color = 'lightgray',
            alpha = 0.5,
            zorder = 2,
            ax = ax[1])

for p in ax[1].patches:
  print("여기 확인 필요:", p, type(p))  # 막대그래프 하나하나가 다 개별 인스턴스 각각 처리해줘야함, height : 요일별 평균
  print('높이', p.get_height(), '너비', p.get_width())
  height = np.round(p.get_height(), 2)
  ax[1].text(p.get_x() + p.get_width()/2, height + 1, height, ha = 'center', size = 12) # p.get_x() x 축 위치, p.get_width()/2 폭의 절반


# 옵션
ax[1].set_title("이렇게 그리고 싶어요", size =16)
ax[1].set_ylim(-2, 25)

plt.show()
```
