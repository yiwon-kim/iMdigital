```python
import yfinance as yf
data = yf.download(['AAPL', 'NVDA'], start='2020-01-01', end = '2024-09-30')
data.head()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(data['Open']['AAPL'], label = 'AAPL', color = 'red')
ax.plot(data['Open']['NVDA'], label = 'NVDA')
ax.legend()
plt.show()
```
