03/02, 03/03 컬럼만 int형으로 변경하세요. 

```python
import pandas as pd

# 데이터프레임 생성
data = [
    ["1,000", "1,100", '1,510'],
    ["1,410", "1,420", '1,790'],
    ["850", "900", '1,185'],
]
columns = ["03/02", "03/03", "03/04"]
df = pd.DataFrame(data=data, columns=columns)
df.info()
```

```python
def com_ma(x):
    return int(x.replace(",",""))
   

result = df[['03/02', '03/03']].map(com_ma) # apply = map
result
```

```python
# 단계별로 처리하기

# 쉼표를 제거하고 int로 변환하는 함수 정의

x = '1,000'
int(x.replace(",", ""))
```

```python
# def로 함수 지정 하고, 작동하는지 여부 확인 후 DF에 적용
def rm_comma(x):
    return int(x.replace(",", ""))

y = '1,000'
print(rm_comma(y))
```

```python
df['03/02'] = df['03/02'].map(com_ma)
df['03/03'] = df['03/03'].map(com_ma)
df.info()
```
